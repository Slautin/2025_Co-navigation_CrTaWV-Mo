```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

from SciFiReaders import IgorIBWReader

import aespm.tools as at

import os
```

```python
def surface_roughness(height_map):
    """
    Compute roughness parameters from a 2D height map.

    Parameters:
        height_map (2D array): surface height values

    Returns:
        dict: Ra, RMS(Rq), Rpv, Rz
    """
    h = np.array(height_map)

    # Remove NaNs if present
    h = h[~np.isnan(h)]

    Ra  = np.mean(np.abs(h - np.mean(h)))          # Arithmetic roughness
    Rq  = np.sqrt(np.mean((h - np.mean(h))**2))    # RMS roughness
    Rpv = np.max(h) - np.min(h)                    # Peak-to-valley
    Rz  = np.mean(np.partition(h, -5)[-5:] - np.partition(h, 5)[:5])  # average peak-valley, ISO style

    return {"Ra": Ra, "Rq": Rq, "Rpv": Rpv, "Rz": Rz}
```

```python
fnames, ibw_data = at.display_ibw_folder(folder="/content/CTWWV-Mo", mode=None, key=None)
```

```python
#make_dict
x, y = [], []
res_dict = {}
for  i in range(len(ibw_data)):
  if i not in [83, 50, 51]:
    res_dict[str(ibw_data[i].header['StagePosX'])] = ibw_data[i].z
```

```python
roughness_array = []
for key in res_dict:
  r_dict = surface_roughness(res_dict[key])
  roughness_array.append([float(key), r_dict["Rq"]*1e9])

roughness_array = np.array(roughness_array)

roughness_array[:,0] = 0.15+(1 - (roughness_array[:,0] -  np.min(roughness_array[:,0]))/ np.ptp(roughness_array[:,0]))*0.7
```

```python
np.savez('CTWWV-Mo_min.npz',**res_dict)
```

```python
fig = plt.figure(figsize=(16, 4))

plt.plot(*roughness_array.T, 'ko-', )
plt.ylabel('Roughness, nm', fontsize=12)
plt.xlabel("(CrtaWV)x - Mo(1-x)", fontsize=12)
plt.grid()
```
